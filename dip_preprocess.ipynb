```python
import pandas as pd
import shutil
import os

# Path to the CSV file containing filenames
csv_file = 'covidx/covidxtestnegative.csv'

# Path to the source folder containing files to be moved
source_folder = 'covidx/test'

# Path to the destination folder where files will be moved
destination_folder = 'covidx/test_positive'

# Column name containing filenames in the CSV file
filename_column = 'filename'  # Change this to the actual column name

# Read the CSV file containing filenames
df = pd.read_csv(csv_file)

# Check if the filename column exists in the dataframe
if filename_column not in df.columns:
    print(f"Error: Column '{filename_column}' not found in the CSV file.")
    exit()

# Iterate over each filename in the CSV file
for index, row in df.iterrows():
    filename = row[filename_column]

    # Construct the source and destination paths
    source_path = os.path.join(source_folder, filename)
    destination_path = os.path.join(destination_folder, filename)

    # Move the file from source folder to destination folder
    shutil.move(source_path, destination_path)
    print(f"Moved {filename} to {destination_folder}")

print("All files moved successfully.")
```

```python
#####Canny #######

import pandas as pd
import shutil
import os
import imageio
from scipy import ndimage
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage import color
from scipy import ndimage

source_folder = 'C:/Users/Priyam.Pandey/Downloads/sars-ct-covid/non-COVID'
files = os.listdir(source_folder)
print(files)
output_folder='C:/Users/Priyam.Pandey/Downloads/sars-ct-covid/norm_canny'
for file_name in os.listdir(source_folder):
    # Check if the file is an image
    if file_name.endswith(('.jpg', '.jpeg', '.png', '.bmp')):
        # Read the image
        image_path = os.path.join(source_folder, file_name)
        image = cv2.imread(image_path)

        # Convert the image to grayscale
        xray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        fourier_gaussian = ndimage.fourier_gaussian(xray_image, sigma=0.05)

        x_prewitt = ndimage.prewitt(fourier_gaussian, axis=0)
        y_prewitt = ndimage.prewitt(fourier_gaussian, axis=1)

        xray_image_canny = np.hypot(x_prewitt, y_prewitt)

        xray_image_canny *= 255.0 / np.max(xray_image_canny)
        output_file = output_folder + f"/{file_name}.png"
        plt.imsave(output_file, xray_image_canny, cmap='terrain')
```

```python
#Adaptive Histogram equalization
import cv2
import matplotlib.pyplot as plt

def apply_clahe(image_path, clip_limit=2.0, tile_grid_size=(8,8)):
    """
    Apply CLAHE (Contrast Limited Adaptive Histogram Equalization) to an image.

    Parameters:
        image_path (str): Path to the input image.
        clip_limit (float): Threshold for contrast limiting.
        tile_grid_size (tuple): Size of the grid for the histogram equalization.

    Returns:
        numpy.ndarray: The image after applying CLAHE.
    """
    # Read the image
    #img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Read as grayscale

    # Create a CLAHE object
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=tile_grid_size)

    # Apply CLAHE to the grayscale image
    clahe_img = clahe.apply(image_path)

    return clahe_img

# Path to your image
source_folder = 'C:/Users/Priyam.Pandey/Downloads/covidx/test_positive'
files = os.listdir(source_folder)
output_folder='C:/Users/Priyam.Pandey/Downloads/covidx/test_positive/covid_clahe'
i=1
for file_name in os.listdir(source_folder):
    # Check if the file is an image
    if file_name.endswith(('.jpg', '.jpeg', '.png', '.bmp')):
        # Read the image
        image_path = os.path.join(source_folder, file_name)
        image = cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)

        # Apply the CLAHE function
        enhanced_img = apply_clahe(image)

        output_file = output_folder + f"/{file_name}.png"
        plt.imsave(output_file, enhanced_img, cmap='gray')
        i=i+1
```
